## Color Threshold for The White Line

```python
def white(img, white_thresh = (235,250)):
   
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS) 

    h_channel, l_channel, s_channel = cv2.split(hls)
   
    #TODO: masking
    ret, binary = cv2.threshold(l_channel,white_thresh[0],white_thresh[1],cv2.THRESH_BINARY)
     
    return binary
```

## Color Threshold for The Yellow Line

```python
def yellow(img, yellow_thresh = (220,255)):
    
    r_channel, g_channel, b_channel = cv2.split(img)
   
    #TODO: masking
                
    ret, binary = cv2.threshold(r_channel,yellow_thresh[0],yellow_thresh[1],cv2.THRESH_BINARY)
     
    return binary
```

## Combined Threshold

```python
task_image = mpimg.imread('./test5.jpg')
binary_white = white(task_image, (235,250))
binary_yellow = yellow(task_image, (230,250))

#TODO: combine both the white and yellow binaries in combined_binary
binary_combined = np.zeros_like(binary_yellow)

binary_combined = cv2.add(binary_white, binary_yellow)



# visualize
f, (ax1,ax2,ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.set_title('1st channel', fontsize=20)
ax1.imshow(binary_white,cmap='gray')
ax2.set_title('2nd channel', fontsize=20)
ax2.imshow(binary_yellow,cmap='gray')
ax3.set_title('combined', fontsize=20)
ax3.imshow(binary_combined,cmap='gray')
```
